```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(i, p = 0.1, c = 1000, m = 5): return max(p * abs(c - i), m)
def g(i, p = 0.07): return max(p*i, 1)
def t(i): return 30

def plot():
    plt.ylim([0, 220])
    plt.xlim([3000, 0])

    a = {i: f(i) for i in range(0, 3000)}
    b = {i: g(i) for i in range(0, 3000)}
    c = {i: t(i) for i in range(0, 3000)}
    for data, label in zip([c, b, a], [ "Constant Step (30)", "Percentage (7%)", "Percentage2 (10%)"]):
        x = list(data.keys())
        y = list(data.values())
        plt.plot(x, y, label=label)

    plt.axvline(1000, color = "red", linewidth=1, linestyle="--")
    plt.legend()
    plt.show()
```

```python
plot()
```

```python
plot({i: f(i) for i in range(0, 3000)})
```

```python
plot({i: g(i) for i in range(0, 3000)})
```

```python
plot({i: t(i) for i in range(0, 3000)})
```
